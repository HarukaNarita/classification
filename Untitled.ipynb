```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
from classification import PolymorphClassifiy
```

```python
PC_1007 = PolymorphClassifiy(roi_path=r"E:\TIRF\211007\rois.csv",
                             image_path=r"E:\TIRF\211007\Stack_m.tif")
```

```python
PC_1007.show_image()
```

```python
PC_1007.plot_contribution_ratio()
```

```python
data = PC_1007.pca_data
feature = PC_1007.feature

db = DBSCAN(eps=0.8, min_samples=10 ).fit(data)
labels = db.labels_
# fig = plt.figure(figsize=(6, 6))
# sns.scatterplot(feature[:,0], feature[:,1], hue=[f"cluster-{x}" for x in labels])

color_names = ["blue","orange"]
colors = [color_names[x] for x in labels]
plt.figure(figsize=(20, 20))
c = 0
l = []
for x, y in zip(feature[:, 0], feature[:, 1]):
    if colors[c] == 'orange':
        plt.text(x, y, str(c), alpha=0.8, size=20)
        l.append(c)
    c += 1
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, color=colors)
print(l)
```

```python
kmeans_model = KMeans(n_clusters=4, random_state=10).fit(feature)
labels = kmeans_model.labels_
color_codes = {0:'#00FF00', 1:'#FF0000', 2:'#0000FF',3:'#00FFFF'}  # green, red, blue
colors = [color_codes[x] for x in labels]

plt.figure(figsize=(20, 20))
c = 0
l = []
for x, y in zip(feature[:, 0], feature[:, 1]):
    if colors[c] == '#00FFFF':
        plt.text(x, y, str(c), alpha=0.8, size=20)
        l.append(c)
    c += 1
plt.scatter(feature[:, 0], feature[:, 1], alpha=0.8, color=colors)
print(l)
# plt.xlim(2,5)
plt.title("Principal Component Analysis")
plt.xlabel("The first principal component score")
plt.ylabel("The second principal component score")
plt.show()
```
